```python
from ai.agents import get_document_agent
from langgraph.checkpoint.memory import InMemorySaver

checkpointer = InMemorySaver()
```

```python
agent = get_document_agent(checkpointer)
```

```python
import uuid

config = {"configurable": {
    "user_id": '2',
    "thread_id": uuid.uuid4(),
    }
}
```

```python
res = agent.invoke(
    {"messages": [{
        "role": 'user',
        "content": "i wanted direct_doc1 give me its content",
    }
    ]},
    config,
)
res
pprint(res['messages'][-1].content)
```
